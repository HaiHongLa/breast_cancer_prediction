```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.drop('id',axis=1)
```

```python
df.describe()
```

```python
def parse_diagnosis(diagnosis):
    if diagnosis == 'B':
        return 0
    else:
        return 1
```

```python
df['diagnosis'] = df['diagnosis'].apply(parse_diagnosis)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop('Unnamed: 32', axis=1)
```

```python
df.info()
```

```python
sns.boxplot(x='diagnosis', y='radius_mean',data=df)
```

```python
plt.figure(figsize=(8,4),dpi=200)
plt.ylabel('Texture Mean')
plt.xlabel('Radius Mean')
plt.legend().get_text()[0].set_text('Benign')
plt.legend().get_text()[1].set_text('Malignant')
sns.scatterplot(x='radius_mean', y='texture_mean',data=df,hue='diagnosis',alpha=0.7)
```
